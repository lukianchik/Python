```python
def from_list_to_string(s):
    stroka = ''
    for i in s:
        stroka += i + ' '
    return stroka

def reverse_string(s):
    s_list = s.split()
    s_list = list(reversed(s_list))
    return from_list_to_string(s_list)

s = str(input(" ,   "))
print(" : " + s)
print(": " + reverse_string(s))
```

```python
def mean(*nums):
    sum = 0
    count = 0
    for n in nums:
        sum += n
        count += 1
    return sum / count

print(mean(1, 2, 3))
print(mean(497825, 2, 3))
print(mean(10, 50, 40, 50, 572))
```

```python
def even(li):
    li2 = []
    for i in li:
        if i % 2 == 0:
            li2.append(i)
    return tuple(li2)


li = [1, 2, 4, 5, 6, 10, 12, 256]
li2 = [i for i in range(1, 100, 5)]
print(even(li))
print(even(li2))
```

```python
from datetime import datetime
from translate import Translator

def month(i):
    test_date = datetime(2023, i, 1)
    res = test_date.strftime("%B")
    return(res)


i = int(input("  : "))
translator= Translator(from_lang="English",to_lang="russian")
month1 = month(i)
perevod = translator.translate(month1)
print(str(i) + ' => ' + perevod.capitalize())
```
