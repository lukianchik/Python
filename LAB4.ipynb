```python
class Book:
    def __init__(self, name, author, pages, price, year, pictures):
        self.name = name
        self.author = author
        self.pages = int(pages)
        self.price = int(price)
        self.year = int(year)
        self.pictures = bool(pictures)

    #Функция вычисления средней стоимости одной страницы
    def price_on_pages(self, name, pages, price):
        mean_for_page = price / pages
        return f'Средняя стоимость одной страницы для {self.name} равна {mean_for_page}'
    
    #Функция увелечения цены в два раза, если название начинаеся с "Программирование"
    def upper_price(self, name, price):
        s = (self.name).split()
        if s[0] == 'Программирование':
            self.price = price * 2
        else:
            self.price = price * 1
        return f'Функция изменения цены для "{self.name}" выполнена'
    
    #Функция выдачи информации о картинках в книге
    def pictures_in_book(self, pictures):
        if self.pictures:
            return 'Картинки в книге есть'
        else:
            return 'Картинок в книге нет'
        
#Класс журнал наследует все из родителя книги, добавляя частоту выпуска
class Journal(Book):
    def __init__(self, name, author, pages, price, year, pictures, frequency):
        Book.__init__(self, name, author, pages, price, year, pictures)
        self.frequency = frequency

#Класс худлжественная книга наследует все из родителя книги, добавляя жанр книги
class Art_book(Book):
    def __init__(self, name, author, pages, price, year, pictures, genre):
        Book.__init__(self, name, author, pages, price, year, pictures)
        self.genre = genre

book1 = Art_book('Преступление и наказание', 'Ф.M Достоевский', 574, 300, 1886, True, 'Роман')
book2 = Book('Программирование для чайников', 'B.И Гольц', 1200, 2500, 2010, False)
journal1 = Journal('TIMES', '', 50, 500, 1935, True, 'once a month')

print(book2.price)
print(book2.upper_price(book2.name, book2.price))
print(book2.price)
print()
print(journal1.price_on_pages(journal1.name, journal1.pages, journal1.price))
print()
print(book2.pictures_in_book(book2.pictures))
```
