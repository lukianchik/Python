```python
import pickle

class InvalidYearError(Exception):
    pass

# Сохранение
def save_data(data, filename):
    try:
        
        with open(filename, 'wb') as file:
            pickle.dump(data, file)
        print("Данные успешно сохранены в файл:", filename)
        
    except IOError:
        print("Ошибка при сохранении данных в файл:", filename)

# Восстановление
def load_data(filename):
    try:

        with open(filename, 'rb') as file:
            data = pickle.load(file)
        print("Данные успешно восстановлены из файла:", filename)
        return data
    
    except FileNotFoundError:
        print("Файл не найден:", filename)

    except IOError:
        print("Ошибка при чтении данных из файла:", filename)

    except pickle.UnpicklingError:
        print("Ошибка при десериализации данных из файла:", filename)

filename = "data.bin"
data = load_data(filename)
if data is None:
    print("Создание нового файла.")

    try:
        n = int(input("Введите год: "))

        if n < 0:
            raise InvalidYearError("Год должен быть положительным числом.")

        if (n % 4 == 0 and n % 100 != 0) or  n % 400 == 0:
            print("Год " + str(n) + " високосный")
            data = {"год": n, "високосный": True}

        else:
            print("Год " + str(n) + " не високосный")
            data = {"год": n, "високосный": False}

        save_data(data, filename)

    #При вводе не int
    except ValueError:
        print("Некорректный ввод года.")

    #При вводе отрицательного года
    except InvalidYearError as e:
        print("Ошибка:", str(e))
```

6. unit-тест - это тестирование отдельных компонентов (функций, методов, классов) кода для проверки их корректности независимо от других компонентов.
