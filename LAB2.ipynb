```python
li = []

with open('input1.txt', 'r') as file:
    for line in file:
        li.append(int(line))


minim = min(li)
maxim = max(li)
print("Числа из файла:", *li)
print("Минимальное число: ", minim)
print("Максимальное число: ", maxim)

message1 = "Минимальное число: " + str(minim) + '\n'
message2 = "Максимальное число: " + str(maxim)
with open('output1.txt', 'a') as file:
    file.write(str(message1))
    file.write(str(message2))
```

```python
with open('input_1.txt', 'r', encoding = 'UTF-8') as file:
    line = file.read()
    li1 = line.split()

with open('input_2.txt', 'r', encoding = 'UTF-8') as file:
    line = file.read()
    li2 = line.split()


with open('output.txt', 'a') as file:
    for i in li1:
        if i in li2:
            print(i)
            file.write(i + '\n')
```

```python
#читаем файл
with open('input4_1.txt', 'r', encoding = 'UTF-8') as file:
    line = file.read()
    line = ''.join([i for i in line if i.isalpha()])
    line = list(line.lower())

#делаем алфавит с ё
a = ord('а')
a = ''.join([chr(i) for i in range(a,a+32)]) + 'ё'

#создаем словарь, где ключ буква, значение количество таких букв
data = {}
for i in a:
    data[str(i)] = int(line.count(i))

#сортируем и записываем в файл
sort_data = sorted(data.items(), key=lambda x: x[1], reverse=True)
with open('output4.txt', 'a', encoding = 'UTF-8') as file:
    for j in sort_data:
        a = j[0]
        b = str(j[1])
        file.write(a + ' ' + b + '\n')

```

1. Для работы с файлами есть функция open(file, mode=""), в которой аргумент file принимает путь, а mode режим работы,                               
   но для более удобного пользования есть метод with open("examp.le", "w") as f:,                       
   который автоматический закрывает файл и бонусом не вызывает исключений при открытии файла(если файла не существует, например)                       
2. readline(): считывает одну строку из файла                            
    read(): считывает все содержимое файла в одну строку                           
    readlines(): считывает все строки файла в список                                  
3. При чтении файла мы можем столкнуться с тем, что у нас разная кодировка. В этом случае мы можем указать кодировку с помощью параметра encoding                             
    with open('output4.txt', 'a', encoding = 'UTF-8') as file: как тут UTF-8 для считывания кириллицы                                    
4. Текстовый файл является читаемым, а бинарный читаемый только для машины,                         
   потому что бинарный файл предоставляется в закодированном виде(такие файлы нужны для чтения изображений, аудио, видеофайлов)                                 
5. Как для дата сайнтистов в csv часто хранят датасеты для работы с ними, поэтому можно сказать, что обычно хранят таблицы                                                         
6. Метод pickle помогает работать со файлами, которые поступают в виде набора байт. Pikcle нельзя использовать для связи с приложениями, которые написаны на других языках.                              
   Десериализация данных pickle, поступающих из ненадежного источника, может                                   
   привести к выполнению произвольного кода.                                        
7. В отличие от JSON, pickle – это протокол, который позволяет сериализовать сложные объекты Python.
