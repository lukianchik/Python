```python
a = input("Введите произвольную строку: ")
a = ''.join([i for i in a if i.isalpha()])
flag = True

for i in a:
    if i.isupper():
        continue
    else:
        flag = False
        break

if flag:
    print("YES")
else:
    print("NO")
```

```python
a = input("Введите свое имя и отчество через пробел: ")

n = a.split()
name = n[0][0]
otches = n[1][0]

print(a + ' - ' + name + '.' + otches + '.')
```

```python
n = input("Введите целые числа через пробел: ")
str_nums = n.split()
int_nums = map(int, str_nums)
print('Уникальные числа: ', *set(int_nums))
```

```python
n = int(input())
if (n % 4 == 0 and n % 100 != 0) or  n % 400 == 0:
    print("Год " + str(n) + " високосный")
else:
    print("Год " + str(n) + " не високосный")
```

```python
n = 1
sum_pos = 0
sum_neg = 0
while n != 0:
    n = int(input("Введите ненулевое число, чтобы добавить его в сумму(при написании 0, закончится ввод): "))
    if n > 0:
        sum_pos += n
    else:
        sum_neg += n

print("Сумма положительных чисел: ", sum_pos)
print("Сумма отрицательных чисел: ", sum_neg)
```
