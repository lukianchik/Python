```python
import urllib
import time

start = time.time()

with open('urls.txt', 'r') as file:
    list_of_url = file.readlines()

#один поток
for i in list_of_url:
    resp = urllib.request.urlopen(i)
    page = resp.read()
    print(page[0 :30])
    print(f'{i} Запрос выполнен успешно')

end = time.time()

print("Время выполнения", (end-start))
```

1. Поток - это сущность исполнения внутри процесса, которая может выполнять инструкции программы. 

2. Процесс - это экземпляр программы, запущенной на компьютере, который выполняется в изолированном адресном пространстве. Ключевое отличие между процессами и потоками заключается в том, что процессы имеют свои собственные независимые адресные пространства, в то время как потоки разделяют адресное пространство процесса.

3. Многопроцессорность - это способность системы обрабатывать несколько процессов (или задач) одновременно. Отличие между многопроцессорностью и многопоточностью заключается в том, что многопроцессорность использует несколько независимых процессоров для одновременной обработки процессов, в то время как многопоточность использует один процессор для параллельного выполнения потоков внутри процесса.

4. Модуль threading из Python предлагает параметры для создания и управления потоками, такие как запуск потока, остановка потока, установка приоритета и т. д.

5. Модуль multiprocessing предоставляет различные варианты API для создания и управления процессами, включая возможность создания подпроцессов, взаимодействие между процессами и т. д.

6. Процесс-демон - это процесс, который работает в фоновом режиме без участия пользователя.

7. GIL - это механизм, используемый в реализации интерпретатора Python, который ограничивает выполнение только одного потока Python в каждый момент времени.

8. Принцип работы GIL состоит в том, что он предотвращает одновременное выполнение нескольких потоков Python за счёт блокировки доступа к внутренним структурам данных интерпретатора.

9. Модуль threading стоит использовать для задач, которые включают в себя асинхронные операции ввода-вывода (I/O), такие как сетевые запросы или операции с файлами, а также для создания responsивных пользовательских интерфейсов.

10. Модуль multiprocessing стоит использовать для задач, требующих параллельной обработки и вычислений, особенно когда необходимо использовать все ядра процессора для ускорения выполнения.

11. Не стоит использовать ни threading, ни multiprocessing в ситуациях, когда задачи не масштабируются параллельно 

12. Основная идея синхронизации потоков при помощи блокировки заключается в использовании механизмов блокировки для предотвращения конфликтов при доступе к общим ресурсам из нескольких потоков.

13. В Python процесс реализации синхронизации потоков с применением блокировки включает использование объектов блокировки из модуля threading, таких как Lock, RLock и Semaphore, для контроля доступа к общим ресурсам.

14. Очередь - это структура данных, представляющая собой упорядоченный набор элементов, куда можно добавлять элементы в конец и извлекать из начала. Основная идея очереди заключается в том, что элементы обрабатываются в порядке их добавления.

15. Очереди широко применяются для совместного программирования в ситуациях, когда несколько процессов или потоков должны обмениваться данными или сообщениями, или когда необходимо организовать асинхронную обработку задач.

16. Основное отличие между обычными очередями и очередями с приоритетами заключается в том, что последние позволяют устанавливать приоритеты для элементов, определяя порядок их обработки на основе заданного приоритета.
